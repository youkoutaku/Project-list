```python
from math import sin
from math import cos
import matplotlib.pyplot as plt

T=0.001
t = []
x_0 = []

for i in range(0,10001):
    i=i*T #0~10000を0~10になるT=0.001
    t.append(i)
    x_0.append( sin(i) )

p1=plt.figure(figsize=(30,15))
p1=plt.plot(t, x_0, '-',color='blue')

p1=plt.hlines([0],t[0], t[-1], color='black', linestyles='dashed', alpha=0.5)
p1=plt.hlines([1],t[0], t[-1], color='black', linestyles='dashed', alpha=0.5)
p1=plt.hlines([-1],t[0], t[-1], color='black', linestyles='dashed', alpha=0.5 )

p1=plt.tick_params(axis='x', labelsize=20)
p1=plt.tick_params(axis='y', labelsize=20)
p1=plt.xlabel("t",fontsize=25)
p1=plt.ylabel("x_0",fontsize=25)
plt.savefig("x_0.png")
```

```python
from math import sin
from math import cos
import matplotlib.pyplot as plt

T=0.001         #サンプリング周期sT
tau = 0.001       #定数tau
A = tau + T/2   #A
B = -tau +T/2   #B
t = []          #描画するためのx軸(時間t)
x_0 = []        #x_0(t)
dx_0 = []       #dx_0(t)
wt = []         #w(t)
xt = []         #x(t)
dxt = []        #dx(t)
y = []          #y(k)

for i in range(0,10001):#0~10000のi
    i=i*T #0~10000を0~10になる.T=0.001
    t.append(i)
    x_0.append( sin(i) )
    dx_0.append( cos(i) )
    wt.append( 0.01*sin(100*i) )
    xt.append( sin(i) + 0.01*sin(100*i))
    dxt.append( cos(i)+ cos(100*i) )

y.append( ( xt[1] - xt[0] )/A )#y[-1]がないので，先にy[0]を求める

for k in range(1,10001):#1~10000のk
    y.append( (xt[k] - xt[k-1] - B*y[k-1] )/A )
    #式でyを求める

p1=plt.figure(figsize=(30,15))
p1=plt.plot(t, y, '-',color='blue')
plt.plot(t, dx_0, '-',color='red')

p1=plt.tick_params(axis='x', labelsize=20)
p1=plt.tick_params(axis='y', labelsize=20)
p1=plt.xlabel("t",fontsize=25)
p1=plt.ylabel("y(k)",fontsize=25)
plt.savefig(f"y(k)_{tau}.png")
```
