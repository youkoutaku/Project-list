```python
# 幾何画像処理　第2回目レポート
#YANG GUANGZE 20T1126N

import numpy as np
import matplotlib.pyplot as plt

P = np.array( [ [10,0],[10,1/3],[1/3,10],[0,10] ] ) #制御点
p = np.array([]) #p(t)関数

px=[]#p(t)関数のx成分
py=[]#p(t)関数のy成分

#ベジェ曲線
for t in range(0,11,1):
    t=t/10
    p = P[0]*((1-t)**3) + P[1]*3*t*((1-t)**2) + P[2]*3*(t**2)*(1-t) + P[3]*(t**3)
    px.append(p[0])
    py.append(p[1])

plt.plot(px,py,'r.')
```

```python
# 幾何画像処理　第2回目レポート
#YANG GUANGZE 20T1126N

import numpy as np
import matplotlib.pyplot as plt

P = np.array( [ [10,0],[10,40*(2**(1/2)-1)/3],[40*(2**(1/2)-1)/3,10],[0,10] ] ) #制御点
p = np.array([]) #p(t)関数

px=[]#p(t)関数のx成分
py=[]#p(t)関数のy成分

#ベジェ曲線
for t in range(0,11,1):
    t=t/10
    p = P[0]*((1-t)**3) + P[1]*3*t*((1-t)**2) + P[2]*3*(t**2)*(1-t) + P[3]*(t**3)
    px.append(p[0])
    py.append(p[1])

plt.plot(px,py,'r.')
```

```python
from math import cos
from math import sin
from math import radians

PX_0 = []
PY_0 = []
#元の円弧
for d in range(0,91,9):#90[deg]の円弧，t=[0,1]で表すと，t=0.1,θ=9[deg]
    PX_0.append(10*cos( radians(d) ))
    PY_0.append(10*sin( radians(d) ) )
    
#青色の円弧
plt.plot(PX_0, PY_0, 'b.')
#赤色のベジェ曲線
plt.plot(px,py,'r.')
```

```python
#t=0.0~1.0のlist生成
t =np.array( range(0,11,1) )
t=t/10

#x座標誤差
er_x = np.array(PX_0)-np.array(px)
#y座標誤差
er_y = np.array(PY_0)-np.array(py)
#誤差
er = (er_x**2+er_y**2)**(1/2)

plt.plot(t,er,'.')#tと誤差の関係図
```
